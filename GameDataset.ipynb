```python
## Importing libs
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
## Reading the file
df_games = pd.read_csv("/content/drive/MyDrive/Python/datasets/games.csv")
df_games.head(5)
```

## Release_year column


```python
#separating year from release date
df_games['release_year'] = pd.to_datetime(df_games['release_date']).dt.year
df_games.head(5)
```

```python
#Converting float to int
df_games['release_year'] = df_games['release_year'].astype(int)
df_games['release_year'].info
```

```python
#locating NaN
nan_rows = df_games[df_games['release_year'].isna()]
nan_rows.head(5)
```

```python

#removendo NaN
df_games['release_year'] = df_games['release_year'].fillna(0).astype(int)
```

## Release_month column

```python
#creating the column
df_games['release_month'] = pd.to_datetime(df_games['release_date']).dt.month
df_games.head(5)
```

```python
#converting to int
df_games['release_month'] = df_games['release_month'].astype(int)
```

```python
#locating nan in months
nan_rows = df_games[df_games['release_month'].isna()]
nan_rows.head(5)
```

```python
#removendo NaN
df_games['release_month'] = df_games['release_month'].fillna(0).astype(int)
df_games.head(5)
```
